<h1>套件</h1>

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime
```

# 讀取資料

```python
stock_data = pd.read_csv("./data/0050.txt", encoding = "cp950")

print(  stock_data.head()  ) ## 查看資料的前五筆
print("--------------------------------------------------------")
print(  stock_data.tail()  ) ## 查看資料的後五筆
```

```python
## 資料抓近五年
stock_data = stock_data.tail(252*5)
```

```python
stock_data
```

```python
stock_data.shape
```

```python
## 將 index 變成 時間格式
stock_data.index = stock_data['年月日'].apply(lambda the_day: datetime.datetime.strptime(str(the_day), "%Y%m%d"))
```

<h1>資料的敘述性統計</h1>

```python
stock_data.describe()
```

# 股價走勢圖

```python
plt.figure(figsize=(15, 6))
plt.plot(stock_data['收盤價'])
```

# 定義簡單迴歸的 y 與 x 

```python
Y = stock_data['收盤價'].values
X = np.arange(len(Y))
```

* 額外補充： np.array 的功能很多，像是可以直接統計運算

```python
print("std  : ", Y.std())   ## (樣本)標準差
print("mean : ", Y.mean())  ## 平均值
print("sum  : ", Y.sum())   ## 全部加總
```

<h1>載入 線性迴歸模型</h1>
<p>資料先進行轉置</p>
<p> .reshape(變成幾個row , 變成幾個col)</p>
<p> 如果輸入 -1 => 資料長度 </p>

```python
Y_reshape = Y.reshape(-1,1)  
X_reshape = X.reshape(-1,1) 
```

```python
X_reshape
```

```python
Y_reshape
```

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()      ### 新的學生 沒計算過迴歸題目的學生
lm.fit(X_reshape, Y_reshape)        ### 迴歸運算
```

<h1>查看迴歸的係數</h1>
<p>y = a + b*x</p>

```python
print("b 斜率：", lm.coef_[0][0])
print("a 截距：", lm.intercept_[0])

print("迴歸公式： y = %.4f + %.4f * x "%( lm.intercept_[0], lm.coef_[0][0]  )) # 每一個 [0]，表示從list 拿第一個元素出來
```

# 趨勢線(迴歸線)

```python
lm_line = np.arange(Y_reshape.size) * lm.coef_[0][0]  + lm.intercept_[0] 
```

```python
plt.figure(figsize=(15, 6))
plt.plot(Y_reshape)
plt.plot(lm_line)
```

# 根據趨勢線(迴歸線)，分別產生 加減1、2 標準差的線

```python
lm_line_add_1_std = lm_line + Y_reshape.std()
lm_line_sub_1_std = lm_line - Y_reshape.std()
lm_line_add_2_std = lm_line + Y_reshape.std() * 2
lm_line_sub_2_std = lm_line - Y_reshape.std() * 2
```

```python
plt.figure(figsize=(15, 6))      ## 圖片大小
plt.plot(Y_reshape, "black") ## 收盤價，  "black" 代表顏色是黑色
plt.plot(lm_line, "brown") ## 趨勢線
plt.plot(lm_line_add_1_std, "g") ## 相對樂觀線
plt.plot(lm_line_sub_1_std, "g") ## 相對悲觀線
plt.plot(lm_line_add_2_std, "r") ## 樂觀線
plt.plot(lm_line_sub_2_std, "r") ## 悲觀線
```
